```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
museum_data = pd.read_csv((r'H:\My Drive\03.Kaggle\04.Data_Visulization\museum_visitors.csv'), index_col="Date",parse_dates=True)
```

```python
museum_data.tail()
```

```python
sns.lineplot(data = museum_data)
```

```python
sns.lineplot(data =museum_data['Avila Adobe'], label = "Avila Adobe")
```

```python
plt.figure(figsize=(10,6))
plt.title("bar plot")
sns.barplot(x =museum_data.index, y =museum_data['Avila Adobe'])
plt.ylabel("Number of Visitors")
```

```python
plt.figure(figsize=(10,6))
plt.title("Heat Map")
sns.heatmap(data = museum_data,annot=False)
```

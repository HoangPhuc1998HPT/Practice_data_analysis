```python

iowa_house_file = r'H:\My Drive\03.Kaggle\02.Machine_Learning\Compertision\home-data-for-ml-course\train.csv'

iowa_house = pd.read_csv(iowa_house_file)
```

```python
y = iowa_house.SalePrice
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
X = iowa_house[features]

X_description = X.describe()
```

```python
print(X_description)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
iowa_house_model = DecisionTreeRegressor(random_state=1)
```

```python
iowa_house_model.fit(X,y)
```

```python
print(X.head())
print(iowa_house_model.predict(X.head()))
```

```python
from sklearn.metrics import mean_absolute_error
predict_iowa_house_price = iowa_house_model.predict(X)
mean_absolute_error(y,predict_iowa_house_price)
```

```python
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
iowa_house_model = DecisionTreeRegressor()
iowa_house_model.fit(train_X,train_y)
val_predictions = iowa_house_model.predict(val_X) 
print(val_predictions)
print(mean_absolute_error(val_y, val_predictions))
```

```python
from sklearn.model_selection import train_test_split
    

# fill in and uncomment
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_iowa_mode = RandomForestRegressor(random_state=1)
rf_iowa_mode.fit(train_X,train_y)
rf_iowa_val_mae = mean_absolute_error(val_y, rf_iowa_mode.predict(val_X))
print( "Validation MAE: {}".format(rf_iowa_val_mae))
```

```python
iowa_house['Predicted_SalePrice'] = rf_iowa_mode.predict(X)

# Display the predicted values along with Id
predictions = iowa_house[['Id', 'Predicted_SalePrice']]
print(predictions)
```

```python
# path to file you will use for predictions
test_data_path = r'H:\My Drive\03.Kaggle\02.Machine_Learning\Compertision\home-data-for-ml-course\test.csv'

# read test data file using pandas
test_data = pd.read_csv(test_data_path)

# create test_X which comes from test_data but includes only the columns you used for prediction.
# The list of columns is stored in a variable called features
test_X = test_data[features]


# make predictions which we will submit. 
test_preds = rf_model_on_full_data.predict(test_X)
```

```python
output = pd.DataFrame({'Id': test_data.Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```
